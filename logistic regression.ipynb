```python
def replace(inp):
    Age=inp[0]
    Pclass=inp[1]
    Sex=inp[2]
    if pd.isnull(Age):
        if Sex=='male':
            if Pclass==1:
                return 41
            elif Pclass==2:
                return 31
            else:
                return 27
        elif Pclass==1:
            return 39
        elif Pclass==2:
            return 29
        else:
            return 22
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass', 'Sex']].apply(replace, axis=1)
```

```python
sns.heatmap(data = train.isnull(), yticklabels=False, cmap='viridis',cbar=False)
```

```python
train.drop('Cabin', axis=1, inplace=True)
```

```python
train.head()
```

```python
train.dropna()
```

```python
male=pd.get_dummies(train['Sex'], drop_first=True)
```

```python
embark=pd.get_dummies(train['Embarked'], drop_first=True)
```

```python
embark
```

```python
train=pd.concat([train, male, embark], axis =1)
```

```python
train.head()
```

```python
train=train.drop(train[['PassengerId','Name','Sex','Embarked','Ticket']],axis=1)
```

```python
train.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Parch', data=train, hue='Survived')
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X= train.drop('Survived', axis=1)
y= train['Survived']
```

```python
train.drop('Survived', axis=1).head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
prediction=lg.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
import sklearn.metrics
```

```python
sklearn.metrics.accuracy_score(y_test,prediction)
```

```python
print(classification_report(y_test,prediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```
